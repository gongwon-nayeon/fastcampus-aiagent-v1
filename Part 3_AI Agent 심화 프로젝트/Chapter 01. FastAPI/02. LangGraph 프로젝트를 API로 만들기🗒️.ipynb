> 기존 테스트에 활용한 https://langchain-ai.github.io/langgraph/agents/memory/ 링크의 리다이렉팅 문제로, https://docs.langchain.com/oss/python/langgraph/add-memory 로 변경하였습니다.

```python
import requests

url = "http://127.0.0.1:8000/ai-assist/invoke"
# payload = {"message": "안녕"}
payload = {"message": "https://docs.langchain.com/oss/python/langgraph/add-memory 이 내용을 한문장으로 요약해주세요."}

response = requests.post(url, json=payload)
print("Status Code:", response.status_code)
print("Response :", response.json()['content'])
```

```python
import requests

url = "http://127.0.0.1:8000/ai-assist/stream"
params = {
    "message": "https://docs.langchain.com/oss/python/langgraph/add-memory 한줄 요약 한글로!"
}
headers = {
    "accept": "application/json"
}

with requests.get(url, params=params, headers=headers, stream=True) as response:
    print("Status Code:", response.status_code)
    print("Streaming response:")

    for line in response.iter_lines(decode_unicode=True):
        if line:  # 빈 줄이 아닐 경우 출력
            print(line)
```
